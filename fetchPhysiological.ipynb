```python
import pandas as pd
import numpy as np

import psycopg2
from datetime import datetime, timedelta

%matplotlib inline

con = psycopg2.connect (user = 'postgres',
                password = 'postgres',
                port = '5432',
                host='localhost',
                database = 'inicudb')

cur  = con.cursor()
```

```python
data = pd.read_csv('times_diaper_change.csv')
ids = data.UHID.unique()
ds = pd.DataFrame()
includedPhyiologicalData = 0
import os
# assign directory
directory = 'apollo_files'
for i in ids:
    uhid_str = "apollo_device_" + i + ".csv"
    f = os.path.join(directory, uhid_str)
    print(f)
    df = pd.read_csv(f)
    print(len(df))
    final_df = data[data.UHID == i]
    df['starttime'] = df.apply(lambda x: todatetime(x['starttime']), axis=1)

    for j in range(len(final_df)):
        start_video = pd.to_datetime(final_df.START.iloc[j])
        end_video = pd.to_datetime(final_df.END.iloc[j])
        q = df[(df['starttime']>=start_video) & (df['starttime']<=end_video)]
        if(len(q) > 0):
            start_video = pd.to_datetime(final_df.START.iloc[j]) - timedelta(seconds=600)
            end_video = pd.to_datetime(final_df.END.iloc[j]) + timedelta(seconds=600)
    #       print(start_video ,", ", end_video, ",", df['starttime'])
            q = df[(df['starttime']>=start_video) & (df['starttime']<=end_video)]
            print(final_df.START.iloc[j] , ',', final_df.END.iloc[j], ',', start_video, ",", end_video, ",", i)
            print(len(q))
            includedPhyiologicalData = includedPhyiologicalData + 1
            q['eventid'] = includedPhyiologicalData
            q['label'] = 'Diaper Change'
            q['procedure_start'] = pd.to_datetime(final_df.START.iloc[j])
            q['procedure_end'] = pd.to_datetime(final_df.END.iloc[j])
            q['title'] = final_df.VideoName.iloc[j]
            ds = ds.append(q,ignore_index=True)
print(includedPhyiologicalData)
print(len(data))
ds.to_csv('apollo_prepped_data_diaper_change.csv')
```
