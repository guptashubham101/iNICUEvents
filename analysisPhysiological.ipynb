```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
ds = pd.DataFrame()
tags = pd.DataFrame()
data = pd.read_csv('apollo_prepped_data_diaper_change.csv')
for i in data['eventid'].unique():
    q = data[data['eventid']==i]
    #print(q.head(1),q.procedure_end.iloc[0],i,l)
    q.sort_values('starttime',inplace=True)  
    q.dropna(subset=['spo2'],inplace=True)

    q['rolling_HR'] = q.heartrate.rolling(10).mean()
    q['rolling_Spo2'] = q.spo2.rolling(10).mean()
    q['rolling_HR'].dropna(inplace=True)
    q['rolling_Spo2'].dropna(inplace=True)        
    q['HR_STV'] = q['heartrate'] - q['rolling_HR']
    q['spo2_STV'] = q['spo2'] - q['rolling_Spo2']

    end_video = pd.to_datetime(q.procedure_end.iloc[0]) 
    start_video = pd.to_datetime(q.procedure_start.iloc[0]) 
    pre = q[pd.to_datetime(q['starttime'])<start_video]
    during = q[(pd.to_datetime(q['starttime'])>=start_video) & (pd.to_datetime(q['starttime'])<=end_video)]
    post =  q[pd.to_datetime(q['starttime'])>end_video]
    q['baseline_HR'] = np.nanmean(pre['heartrate'])
    q['baseline_spo2'] = np.nanmean(pre['spo2'])
    q['baseline_HR_STV'] = np.nanvar(pre['HR_STV'])
    q['baseline_spo2_STV'] = np.nanvar(pre['spo2_STV'])
    q['during_HR'] = np.nanmean(during['heartrate'])
    q['during_spo2'] = np.nanmean(during['spo2'])
    q['during_HR_STV'] = np.nanvar(during['HR_STV'])
    q['during_spo2_STV'] = np.nanvar(during['spo2_STV'])
    q['post_HR'] = np.nanmean(post['heartrate'])
    q['post_spo2'] = np.nanmean(post['spo2'])
    q['post_HR_STV'] = np.nanvar(post['HR_STV'])
    q['post_spo2_STV'] = (post['spo2_STV']).var()
    test = q.copy()
    test.drop_duplicates(subset=['title','label'],keep='first',inplace=True)
    tags = tags.append(test,ignore_index=True)                
    ds = ds.append(q,ignore_index=True)
tags.to_csv('tags_diaper_change.csv')
ds.to_csv('data_final_diaper_change.csv')
```
