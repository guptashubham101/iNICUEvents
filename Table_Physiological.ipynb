```python
import pandas as pd
from prettytable import PrettyTable
import scipy
import math
from scipy import special
data = pd.DataFrame()
df = pd.read_csv('tags_diaper_change.csv')
data = data.append(df)
df = pd.read_csv('tags_feeding.csv')
data = data.append(df)
df = pd.read_csv('tags_patting.csv')
data = data.append(df)
```

```python
uhidlessthan32weeks = ['RSHI.0000032084','RSHI.0000032083','RSHI.0000031101','RSHI.0000031430','RSHI.0000031843','RSHI.0000031177','RSHI.0000021833','RSHI.0000023451','191001977']
uhidmorethan32weeks = ['RSHI.0000032831','RSHI.0000030088','RSHI.0000032830','RSHI.0000032466','RSHI.0000024482','200100035','RSHI.0000030046','200100147','200101135','RSHI.0000029568']
uhidlessthan32tags = pd.DataFrame()
for uhid in uhidlessthan32weeks:
    ds = data[data['uhid'] == uhid]
    uhidlessthan32tags = uhidlessthan32tags.append(ds)

uhidmorethan32tags = pd.DataFrame()
for uhid in uhidmorethan32weeks:
    ds = data[data['uhid'] == uhid]
    uhidmorethan32tags = uhidmorethan32tags.append(ds)
```

```python
y = uhidlessthan32tags[(uhidlessthan32tags['label']=='Feeding')|(uhidlessthan32tags['label']=='Patting')|(uhidlessthan32tags['label']=='Diaper Change')]


labels = ['Patting','Feeding','Diaper Change']

table = PrettyTable(['Manipulations','Parameter','Baseline','During','Post','p-values during','p-values post'])




for i in labels:
    answer = y[y['baseline_HR'] > 1.000]
    tags = answer[answer['label']==i]

    variance = sum([((x - tags.baseline_HR.mean()) ** 2) for x in tags.baseline_HR]) / len(tags) 
    res_d_baseline = variance ** 0.5
    
    initalMean = tags.baseline_HR.mean()
    
    answer = y[y['during_HR'] > 1.000]
    tags = answer[answer['label']==i]

    variance = sum([((x - tags.during_HR.mean()) ** 2) for x in tags.during_HR]) / len(tags) 
    res_d_during = variance ** 0.5
    
    zscore = (initalMean-tags.during_HR.mean()) / (res_d_during/math.sqrt(len(tags)))
    if(zscore >= 0):
        zscore = -(zscore)
    p_value_during = scipy.special.ndtr(zscore)
    
    initalMean = tags.during_HR.mean()
    
    answer = y[y['post_HR'] > 1.000]
    tags = answer[answer['label']==i]
    
    variance = sum([((x - tags.post_HR.mean()) ** 2) for x in tags.post_HR]) / len(tags) 
    res_d_post = variance ** 0.5
    
    zscore = (initalMean-tags.post_HR.mean()) / (res_d_post/math.sqrt(len(tags)))
    if(zscore >= 0):
        zscore = -(zscore)
    p_value_post = scipy.special.ndtr(zscore)

    table.add_row([i,'HR',str(round(tags.baseline_HR.mean(),1)) + '(' + str(round(tags.baseline_HR.median(),1)) + ',' + str(round(res_d_baseline,1)) + ')',str(round(tags.during_HR.mean(),1)) + '(' + str(round(tags.during_HR.median(),1)) + ',' + str(round(res_d_during,1)) + ')',str(round(tags.post_HR.mean(),1)) + '(' + str(round(tags.post_HR.median(),1)) + ',' + str(round(res_d_post,1)) + ')',round(p_value_during,2),round(p_value_post,2)])
    
    
    
    
    
    answer = y[y['baseline_spo2'] > 1.000]
    tags = answer[answer['label']==i]

    variance = sum([((x - tags.baseline_spo2.mean()) ** 2) for x in tags.baseline_spo2]) / len(tags) 
    res_d_baseline = variance ** 0.5
    
    initalMean = tags.baseline_spo2.mean()
    
    answer = y[y['during_spo2'] > 1.000]
    tags = answer[answer['label']==i]

    variance = sum([((x - tags.during_spo2.mean()) ** 2) for x in tags.during_spo2]) / len(tags) 
    res_d_during = variance ** 0.5
    
    zscore = (initalMean-tags.during_spo2.mean()) / (res_d_during/math.sqrt(len(tags)))
    if(zscore >= 0):
        zscore = -(zscore)
    p_value_during = scipy.special.ndtr(zscore)
    
    initalMean = tags.during_spo2.mean()
    
    answer = y[y['post_spo2'] > 1.000]
    tags = answer[answer['label']==i]
    
    variance = sum([((x - tags.post_spo2.mean()) ** 2) for x in tags.post_spo2]) / len(tags) 
    res_d_post = variance ** 0.5
    
    zscore = (initalMean-tags.post_spo2.mean()) / (res_d_post/math.sqrt(len(tags)))
    if(zscore >= 0):
        zscore = -(zscore)
    p_value_post = scipy.special.ndtr(zscore)

    table.add_row([i,'Spo2',str(round(tags.baseline_spo2.mean(),1)) + '(' + str(round(tags.baseline_spo2.median(),1)) + ',' + str(round(res_d_baseline,1)) + ')',str(round(tags.during_spo2.mean(),1)) + '(' + str(round(tags.during_spo2.median(),1)) + ',' + str(round(res_d_during,1)) + ')',str(round(tags.post_spo2.mean(),1)) + '(' + str(round(tags.post_spo2.median(),1)) + ',' + str(round(res_d_post,1)) + ')',round(p_value_during,2),round(p_value_post,2)])

print(table)
```
