```python
def extract_actual_time_neo(x):
    try:
        print(x)
        if("videoOnly_" in x):
            t = str(x).split("videoOnly_")[1]
        elif("NEOTINY116_" in x):
            t = str(x).split("NEOTINY116_")[1]
        elif("NEOTINY114_" in x):
            t = str(x).split("NEOTINY114_")[1]
        elif("NEOTINY113_" in x):
            t = str(x).split("NEOTINY113_")[1]
        elif("NEOTINY118_" in x):
            t = str(x).split("NEOTINY118_")[1]
        elif("NEOTINY126_" in x):
            t = str(x).split("NEOTINY126_")[1]
        elif("720p_" in x):
            t = str(x).split("720p_")[1]
        elif("NEO109.stream_" in x):
            t = str(x).split("NEO109.stream_")[1]
        elif("NEOTINY104.stream_" in x):
            t = str(x).split("NEOTINY104.stream_")[1]
        elif("NEO102.stream_" in x):
            t = str(x).split("NEO102.stream_")[1]
        elif("NEO103.stream_" in x):
            t = str(x).split("NEO103.stream_")[1]
        elif("NEO106.stream_" in x):
            t = str(x).split("NEO106.stream_")[1]
        elif("NEO101.stream_" in x):
            t = str(x).split("NEO101.stream_")[1]
        else:
            return
        if t != '720p':
            from_zone = tz.tzutc()
            to_zone = tz.tzlocal()
            test = ((t.split("-")[3]).split(".")[0] +":"+ (t.split("-")[3]).split(".")[1] + ":"+ (t.split("-")[3]).split(".")[2])
            if("IST" in t):
                indexStr = t.split("-IST_")[1]
            elif("UTC" in t):
                indexStr = t.split("-UTC_")[1]
            index = int(indexStr.split(".mp4")[0])
            if(index == 0):
                return test
            else:
                initialTime = int((t.split("-")[3]).split(".")[0])

                while(index != 0):
                    initialTime = initialTime + 1
                    if(initialTime == 24):
                        initialTime = 0
                    index = index - 1
                if(len(str(initialTime)) == 1):
                    initialTime = "0" + str(initialTime)
                return str(initialTime) + ":00:00"
        else:
            t = str(x).split("_")[2]
            from_zone = tz.tzutc()
            to_zone = tz.tzlocal()
            print(t)
            test = ((t.split("-")[3]).split(".")[0] +":"+ (t.split("-")[3]).split(".")[1] + ":"+ (t.split("-")[3]).split(".")[2])
            if("IST" in t):
                indexStr = t.split("-IST_")[1]
            elif("UTC" in t):
                indexStr = t.split("-UTC_")[1]
            index = int(indexStr.split(".mp4")[0])

            if(index == 0):
                return test
            else:
                initialTime = int((t.split("-")[3]).split(".")[0])
                while(index != 0):
                    initialTime = initialTime + 1
                    if(initialTime == 24):
                        initialTime = 0
                    index = index - 1
                if(len(str(initialTime)) == 1):
                    initialTime = "0" + str(initialTime)
                return str(initialTime) + ":00:00"
    except Exception as e:
        print(e)
        return
def extract_time(x):
    try:
        print(x)
        if("videoOnly_" in x):
            t = str(x).split("videoOnly_")[1]
        elif("NEOTINY116_" in x):
            t = str(x).split("NEOTINY116_")[1]
        elif("NEOTINY114_" in x):
            t = str(x).split("NEOTINY114_")[1]
        elif("NEOTINY113_" in x):
            t = str(x).split("NEOTINY113_")[1]
        elif("NEOTINY118_" in x):
            t = str(x).split("NEOTINY118_")[1]
        elif("NEOTINY126_" in x):
            t = str(x).split("NEOTINY126_")[1]
        elif("720p_" in x):
            t = str(x).split("720p_")[1]
        elif("NEO109.stream_" in x):
            t = str(x).split("NEO109.stream_")[1]
        elif("NEOTINY104.stream_" in x):
            t = str(x).split("NEOTINY104.stream_")[1]
        elif("NEO102.stream_" in x):
            t = str(x).split("NEO102.stream_")[1]
        elif("NEO103.stream_" in x):
            t = str(x).split("NEO103.stream_")[1]
        elif("NEO106.stream_" in x):
            t = str(x).split("NEO106.stream_")[1]
        elif("NEO101.stream_" in x):
            t = str(x).split("NEO101.stream_")[1]
        else:
            return
        if t != '720p':
            from_zone = tz.tzutc()
            to_zone = tz.tzlocal()
            test = t.split("-")[0] +"-" + t.split("-")[1] + "-" + t.split("-")[2] + " " + ((t.split("-")[3]).split(".")[0] +":"+ (t.split("-")[3]).split(".")[1] + ":"+ (t.split("-")[3]).split(".")[2])
            utc = datetime.strptime(test, '%Y-%m-%d %H:%M:%S')
            central = utc.astimezone(to_zone)
            central = pd.to_datetime(central)
        else:
            t = str(x).split("_")[2]
            from_zone = tz.tzutc()
            to_zone = tz.tzlocal()
            test = t.split("-")[0] +"-" + t.split("-")[1] + "-" + t.split("-")[2] + " " + ((t.split("-")[3]).split(".")[0] +":"+ (t.split("-")[3]).split(".")[1] + ":"+ (t.split("-")[3]).split(".")[2])
            utc = datetime.strptime(test, '%Y-%m-%d %H:%M:%S')
            central = utc.astimezone(to_zone)
            central = pd.to_datetime(central)
    except Exception as e:
        print(e)
        return
    return ((str(pd.to_datetime(central)).split("+")[0]))

def extract_neo(x):
    s = str(x).split("_")[0]
    if "RSHI" in s:
        s = str(x).split("_")[1]
    print(s)
    return (s.split(" ")[-1].split(".")[0].lower())

def extract_seq(x):
    s = str(x).split("_")[-1]
    
    return s.split(".")[0]

def to_split_0(x):
    x = str(x)
    return x.split(".")[0]
def to_split_1(x):
    try:
        x = str(x)
        return x.split(".")[1]
    except:
        pass

def date_extract(x):
    try:
        return x.split(" ")[0]
    except:
        return

def calculate_end_date(x,y):
    try:
        x = str(x)
        xarray = x.split(".")
        totalseconds = int(int(xarray[0])*60 + int(xarray[1]))
        central = pd.to_datetime(y) + timedelta(seconds=totalseconds)
        return central
    except:
        return
def time_join(video,x,y):
    try:
        to_zone = tz.tzlocal()
        test = str(x)+" "+str(y)
        utc = datetime.strptime(test, '%Y-%m-%d %H:%M:%S')
        central = utc.astimezone(to_zone)
        if "IST" in video:
            central = pd.to_datetime(central) - timedelta(seconds=19800)
        else:
            central = pd.to_datetime(central)
        return central
    except:
        try:
            to_zone = tz.tzlocal()
            test = str(x)+" "+str(y)
            utc = datetime.strptime(test, '%Y-%m-%d %H.%M.%S')
            central = utc.astimezone(to_zone)
            if "IST" in video:
                central = pd.to_datetime(central) - timedelta(seconds=19800)
            else:
                central = pd.to_datetime(central)
            return central
        except:
            pass
    
```
